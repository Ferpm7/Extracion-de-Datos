```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Filtro por objeto
# Filtrar por el valor en la columna "NOMBRE_VENDEDOR" igual a "LETICIA RAMIREZ HERNANDEZ" PARA VER UN RESUMEN CON LO RELACIONADO A LETICIA, SE OCUPO LA VARIABLE NOMBRE_VENDEDOR
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("FILTRO1.csv")
```

```python
#Filtro por filas
# Seleccionar filas de forma continua desde la fila 3 hasta la fila 9
filtro2= df.iloc[3:10,: ]           #Filas consecutivas
#filtro2=df.iloc[ [8, 24, 100] , : ]  #Filas no consecutivas
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("FILTRO2.csv")
```

```python
#Filtro por columnas
# Seleccionar todas las filas y las columnas de la 1 a la 2 (columnas 2 y 3) SE MUESTRAN LAS VARIABLES FECHA_DOC Y NOMBRE DEL VENDEDOR
filtro3=df.iloc[ : , 1:3]       #Columnas consecutivas
#filtro3=df.iloc[ : , [1,4,6]]  #Columnas no consecutivas
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("FILTRO3.csv")
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Aplicamos Filtro de filas vs columnas
# Filtrar filas donde el valor en la columna "NOMBRE_VENDEDOR" sea "LETICIA RAMIREZ HERNANDEZ" o "DIEGO ARMANDO LEON DURAN", y seleccionar la columna "CANT" PARA VER LA CANTIDAD POR ESTOS 2 NOMBRES
filtro4=df1.loc[["LETICIA RAMIREZ HERNANDEZ","DIEGO ARMANDO LEON DURAN"], ["CANT"]]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("FILTRO4.csv")
```

```python
#Filtro por cabecera
#Esta línea crea un nuevo DataFrame llamado filtro5 que contiene las primeras 5 filas de df. Es una forma rápida de ver las primeras filas de un DataFrame para una vista previa.
filtro5=df.head()
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("FILTRO5.csv")
```

```python
#Filtro por cola
# Esta línea crea un nuevo DataFrame llamado filtro6 que contiene las últimas 5 filas de df. Es una forma rápida de ver las últimas filas de un DataFrame.
filtro6=df.tail()
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("FILTRO6.csv")
```

```python
#Filtro por comparacion
#Esta línea crea un nuevo DataFrame llamado filtro7 que contiene todas las filas de df donde el valor en la columna "PRECIO_UNITARIO" sea mayor que 2000. SE UTILIZO LA VARIABLE PRECIO_UNITARIO PARA VER EL COSTO POR UNIDAD MAYOR A 2000
filtro7=df[df["PRECIO_UNITARIO"] > 2000]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("FILTRO7.csv")
```

```python
#Filtro Y
#Esta línea crea un nuevo DataFrame llamado filtro8 que contiene todas las filas de df donde el valor en la columna "PRECIO_UNITARIO" sea menor que 6000
#y el valor en la columna "FECHA_DOC" sea igual a "2022-03-14". Utiliza el operador & para realizar una operación booleana "y" entre las dos condiciones.
filtro8=df[(df["PRECIO_UNITARIO"] < 6000) & (df["FECHA_DOC"] == "2022-03-14")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("FILTRO8.csv")
```

```python
#Filtro o
#Esta línea crea un nuevo DataFrame llamado filtro9 que contiene todas las filas de df donde el valor en la columna "PRECIO_UNITARIO" sea menor que 6000 o el valor en la columna "FECHA_DOC" sea igual a "2022-03-14".
#Utiliza el operador | para realizar una operación booleana "o" entre las dos condiciones.
filtro9=df[(df["PRECIO_UNITARIO"] < 6000) | (df["FECHA_DOC"] == "2022-03-14")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("FILTRO9.csv")
```

```python
#Filtro not
#Esta línea crea un nuevo DataFrame llamado filtro10 que contiene todas las filas de df donde el valor en la columna "PRECIO_UNITARIO" no sea menor que 6000 y el valor en la columna "FECHA_DOC" no sea igual a "2022-03-14".
# El operador ~ se usa para negar ambas condiciones.
filtro10=df[~(df["PRECIO_UNITARIO"] < 6000) & ~(df["FECHA_DOC"] == "2022-03-14")]
filtro10
```

```python
#Convertir archivo filtrado a CSV
filtro10.to_csv("FILTRO10.csv")
```
